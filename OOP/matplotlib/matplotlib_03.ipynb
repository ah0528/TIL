# pandas로 그래프 그리기

## pandas의 그래프 구조

`Series_data.plot([kind='그래프종류', option])
DataFrame_data.plot([x=label 혹은 position, y=label 혹은 position, kind='그래프종류, option']`

- kind : 그래프의 종류 지정
|kind 옵션|설명|
|---|---|
|line|선 그래프(기본값)|
|scatter|산점도(DataFrame데이터만 가능)|
|bar|막대 그래프|
|barh|가로 막대 그래프|
|hist|히스토그램|
|pie|파이 그래프|


- `Series_data.plt(kind='bar')`은 `Series_data.plot.bar()`처럼 사용 가능
- kind 생략시 선 그래프가 그려짐

## 선 그래프

**Series 데이터를 그래프로 그리기**

```python
import pandas as pd
import matplotlib.pyplot as plt

s1 = pd.Series([1,2,3,4,5,6,7,8])
s1.plot()
plt.show()
```

```python
s2 = pd.Series([1,2,3,4,5,6,7,8], 
               index=['2022-01-22', '2022-01-23', '2022-01-24', '2022-01-25', '2022-01-26', '2022-01-27', '2022-01-28', '2022-01-29'])
s2.plot(grid=True)
plt.show()
# 일반적인 배열 그래프 생성과 달리 grid를 옵션으로 사용해야 함. plt.grid()는 오류발생
```

**DataFrame 데이터를 그래프로 그리기**

```python
df = pd.DataFrame({'a':[4394,5252,7655],'b':[6564,6246,8653],'c':[1334,6345,9672],'d':[3634,8523,7634], 'e':[6321,1246,4374]}, index=['2001','2002','2003'])
df
```

```python
df.plot()
plt.show()
```

```python
df.plot(grid=True, style=['r--*'])
plt.xlabel('year')
plt.title('DataFrame Graph Exercise')

plt.show()
```

```python
df_graph = df.plot(grid=True, style=['r--*'])
df_graph.set_xlabel('year')
df_graph.set_title('DataFrame Graph Exercise')

plt.show()
```

```python
df.plot(grid=True, style=['r--*'])
```

## 산점도

```python
temperature = [25.2, 27.4, 22.9, 26.2, 33.1, 30.4, 36.1, 34.4]
icecream = [15, 18, 16, 24, 35, 40, 37, 29]
dict_data = {'temperature':temperature, 'sells':icecream}
df_ice = pd.DataFrame(dict_data)
df_ice
```

```python
df_ice.plot(kind='scatter', x='temperature', y='sells', grid=True)
```

```python
df_ice.plot.scatter(x='temperature', y='sells', grid=True)

plt.show()
```

## 막대 그래프

```python
df = pd.DataFrame([5,14,13,2], index=['A', 'B', 'C', 'D'], columns=['Student'])
df
```

```python
df_graph = df.plot.bar(grid=True)
df_graph.set_ylabel('grade')

plt.show()
```

## 히스토그램

```python
df.plot.hist(grid=True)

plt.show()
```

## 파이 그래프

```python
fruit = ['apple', 'pineapple', 'melon', 'watermelon', 'mango']
result = [7,3,8,4,9]

df_fruit = pd.Series(result, index=fruit)
df_fruit
```

```python
df_fruit.plot.pie()

plt.show()
```

```python
df_fruit.plot.pie(table=True, explode=[0,0.1,0,0,0], autopct='%.1f%%')

plt.show()
```
