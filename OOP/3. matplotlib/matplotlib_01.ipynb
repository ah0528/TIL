# 데이터 시각화 matplotlib
- 파이썬에서 데이터를 효과적으로 시각화하기 위해 만든 라이브러리
- matplotlib를 이용하면 몇 줄의 코드로 간단하게 2차원 선 그래프(plot), 산점도(scatter plot), 막대 그래프(bar chart), 히스토그램(histogram), 파이 그래프(pie chart) 등을 그릴 수 있음

- matplotlib의 그래프 기능을 사용하려면 matplotlib의 서브모듈을 불러와야 함
`import matplotlib.pyplot as plt`

- 만약 별도의 팝업창에 그래프를 출력하려면 '그래프함수()'를 실행하기 전에 `%matplotlib qt`를 실행해야 함
- 그래프를 다시 코드 결과 출력 부분에 출력하려면 '그래프함수()'를 실행하기 전에 `%matplotlib inline`를 실행해야 함

## 1. 선그래프
### 기본적인 선 그래프 그리기

**plt.plot([x,]y[,fmt])**

- 순서가 있는 숫자 데이터를 시각화하거나 시간에 따라 변하는 숫자 데이터를 시각화하는 데 많이 사용
- x와 y의 시퀀스의 길이가 같아야 함
- x와 y 데이터 중 x는 생략이 가능하며 생략시 0부터 y의 개수 만큼 1씩 증가하는 값으로 자동 할당됨
- fmt가 생략되면 기본 형식으로 그래프를 그림
- 그래프 객체의 정보 없이 그래프만 출력하려면 '그래프함수()'를 실행한 이후에 plt.show()를 실행하면 됨

```python
import matplotlib.pyplot as plt
```

```python
# x가 생략된 선 그래프 그리기
# 그래프 객체 정보없이 그래프만 출력하기
data = [10,14,19,20,25]
plt.plot(data)
plt.show()
```

```python
# 별도의 팝업창에 그래프 나타내기
%matplotlib qt
```

```python
data=[12,15,19,24,26]
plt.plot(data)
# plt.show()
```

```python
# 다시 결과창에 그래프 나타내기
%matplotlib inline
```

```python
# y = 2x**2 함수의 그래프 그리기
import numpy as np
x = np.arange(-4.5, 5, 0.5)
y = 2*x**2
plt.plot(x,y)
plt.show()
```

## 여러 그래프 그리기

#### 하나의 그래프 창에 여러 그래프 그리기
1. 연속해서 plot()을 수행하기

`plt.plot([x1,]y1[,fmt1])
plt.plot([x2,]y2[,fmt2])
plt.plot([x3,]y3[,fmt3])
plt.plot([x4,]y4[,fmt4])`


2. 하나의 plot()에 표시하고자 하는 데이터를 모두 입력하기

`plt.plot(x1, y1, [fmt1,] x2, y2, [fmt2,] x3, y3, [fmt3,] x4, y4[, fmt4]`
  - 이 경우에는 x를 생략할 수 없음(x와 y가 짝을 이뤄야하기 때문)

```python
# 연속해서 plot()을 수행해 여러 그래프 그리기
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2
y2 = 5*x+30
y3 = 4*x**2+10

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.show()
```

```python
# 하나의 plot()에 여러 그래프 그리기
plt.plot(x,y1,x,y2,x,y3)
plt.show()
```

#### 새로운 그래프 창에 그래프 그리기
1. '그래프함수()'를 실행하기 전에 `plt.figure()`을 실행


2. 그래프 창의 번호를 명시적으로 지정한 후 해당 창에 그래프 그리기 : `plt.figure(n)`

- 정수 n을 지정하면 지정된 번호로 그래프 창이 지정되고 이후에 '그래프함수()'를 실행하면 지정된 번호의 그래프 창에 그래프가 그려짐 
- 만약 지정된 번호의 그래프 창이 없다면 새로 그래프 창을 생성한 후에 그래프가 그려짐
- %matplotlib qt를 실행해 별도의 팝업창에서 그래프를 그린다면 제목에 'Figure번호'가 표시됨
- `plt.figure(n)`를 지정한 후에 현재 그래프 창의 그래프를 모두 지우려면 `plt.clf()`를 실행하면 됨
- 현재의 그래프 창을 닫으려면 `plt.close()`를 이용하면 됨

```python
plt.plot(x,y1)

plt.figure()

plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y1 = x**2-2
y2 = 20*np.cos(x)**2

plt.figure(1) #1번 그래프 창 생성
plt.plot(x,y1)


plt.figure(2)
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y2)

plt.figure(2)
plt.clf()
plt.plot(x,y1)

plt.show()
```

#### 하나의 그래프 창을 하위 영역으로 나누어 각 영역에 그래프 그리기

`plt.subplot(m, n, p)`

- 그래프 창을 m x n 행렬로 나누고 p번째 영역을 가르킴 (영역의 개수는 m*n개)

```python
x = np.arange(0,10,0.1)

y1 = 0.3*(x-5)**2+1
y2 = -1.5*x+3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x)+1

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

#### 그래프의 출력 범위 지정하기

`plt.xlim(xmin, xmax)`

`plt.ylim(ymin, ymax)`

- 전체 그래프 중 관심 영역만 확대해서 보고 싶을 때 출력되는 그래프의 x축과 y축의 좌표 범위를 지정해 전체 그래프 중 관심 영역만 그래프로 그릴 수 있음

- `plt.xlim(xmin, xmax)`은 독립적으로 동작하므로 둘 중 하나만 이용할 수도 있음

- `plt.xlim(xmin, xmax)`은 모두 '그래프함수()'로 그래프를 출력한 후에 실행해야 현재 출력한 그래프의 x축과 y축의 범위를 지정하거나 가져올 수 있음

```python
x = np.linspace(-5,5,100)
y1 = x**3
y2 = 10*x**2 -2

plt.plot(x, y1, x, y2)
plt.show()
```

```python
# 확대해서 보기
plt.plot(x,y1,x,y2)
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
```

## 그래프 꾸미기

`fmt = '[color][line_style][marker]'`

- fmt 옵션을 활용해 그래프에서 선의 출력 형식을 지정하고 그래프에 축의 라벨, 제목, 격자, 문자열을 추가할 수 있음

- color, line_style, marker은 하나씩만 지정할 수도, 조합해서 지정할 수도 있음
- 조합시 순서가 고정된 것은 아님

- 선의 스타일 지정을 위한 약어
|선 스타일|설명|
|---|----|
|- (solid)|실선|
|-- (dashed)|파선|
|: (dotted)|점선|
|-. (dashdot)|파선 점선 혼합선|

- 마커 지정을 위한 약어
|마커|설명|
|----|----|
|o|원 모양|
|^, v, <, >|삼각형 위쪽, 아래쪽, 왼쪽, 오른쪽|
|s|사각형|
|p|오각형|
|h, H|육각형1, 육각형2|
| * |별 모양|
|+|더하기|
|x, X|x, 채워진x|
|D, d|다이아몬드, 얇은 다이아몬드|

```python
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c') # 컬러 약어로 컬러 지정함
plt.show()
```

```python
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.') # 선스타일 약어로 선의 스타일 지정함
plt.show()
```

```python
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d') # 마커 약어로 마커 지정함
plt.show()
```

```python
# fmt옵션에 컬러, 선스타일, 마커를 혼합하여 사용하기
plt.plot(x,y1,'r+-',x,y2,'c*--',x,y3,'h:k',x,y4,'d:m')
plt.show()
```

#### 라벨, 제목, 격자, 범례, 문자열 표시
- 라벨 표시 : `plt.xlabel('문자열')`, `plt.ylabel('문자열')`


- 제목 표시 : `plt.title('문자열')`


- 격자 표시 : `plt.grid(True)` 또는 `plt.grid()`

  - 이미 격자가 있는 그래프애서 격자를 제거하고 싶다면 `plt.rid(False)`를 사용
  
  - `grid()`를 수행하면 show()를 수행하지 않아도 그래프 객체의 정보를 출력하지 않고 그래프만 출력함
  
  
- 범례 표시 : `legend(['str1', 'str2', 'str3', ...])`

  - `legend()`에서 범례의 위치를 지정하지 않으면 자동으로 지정되지만, loc 옵션으로 범례의 위치를 지정할 수 있음
    - loc='위치 문자열'
    - loc='위치 코드'
|범례위치|위치 문자열|위치 코드|
|--------|-----------|--------|
|자동 선정|best|0|
|상단 우측|upper right|1|
|상단 좌측|upper left|2|
|하단 좌측|lower left|3|
|하단 우측|lower right|4|
|우측|right|5|
|중앙 좌측|center left|6|
|중앙 우측|center right|7|
|하단 중앙|lower center|8|
|상단 중앙|upper center|9|
|중앙|center|10|


- 문자열 표시 : `plt.text(x, y, '문자열')`

한글을 사용하고 싶다면 matplotlib에서 사용하는 폰트를 한글 폰트로 지정해야 함

```python
x = np.arange(-4.5,5,0.5)
y = 2*x**3

# 라벨 표시하기
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 제목 표시하기
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```

```python
# 그리드 표시하기
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.grid()    # plt.grid(True)와 동일
```

```python
# 범례 표시하기
y2 = 5*x + 10

plt.plot(x,y, x, y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.grid()
plt.legend(['data01','data02'])
plt.show()
```

```python
# 범례 표시하기 - 우측 하단에
y2 = 5*x + 10

plt.plot(x,y, x, y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.grid()
plt.legend(['data01','data02'], loc='lower right')    # loc = 4 와 동일
plt.show()
```

```python
# 문자열 표시하기
y2 = 5*x + 10

plt.plot(x,y, x, y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.grid()
plt.legend(['data01','data02'], loc='lower right')
plt.text(0,100,'Have a Good Day:)')
plt.show()
```
