## 2. 산점도
두 개의 요소로 이뤄진 데이터 집합의 관계를 시각화

`plt.scatter(x, y [,s=size_n, c=colors, marker='marker_string', alpha=alpha_f])`

- s 옵션에 하나의 숫자만 입력하면 모든 마커에 동일한 크기를 적용, 배열이나 시퀀스로 입력하면 마커마다 크기를 다르게 적용

- c 옵션도 하나의 숫자만 입력하면 모든 마커에 동일한 색상을 적용, 시퀀스로 입력하면 마커마다 색상을 다르게 적용

- alpha 옵션은 투명도를 지정하는데, 0~1 범위의 실수를 입력. 0은 완전 투명, 1은 완전 불투명.

```python
# 키와 몸무게의 데이터 쌍을 산점도로 그리기
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter(height, weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid()
```

```python
# 마커에 빨간색을 주고, 크기는 500으로 그리기

plt.scatter(height, weight, c ='r', s = 500)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid()
```

```python
# 마커의 크기와 컬러를 다 다르게 지정하기
import numpy as np

size= 100 * np.arange(1,8)

plt.scatter(height, weight, s=size, c=['r', 'y', 'k', 'g', 'b', 'c', 'm'])
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid()
```

```python
# 지역별 인구밀도 산점도로 그리기

city = ['Seoul','Incheon','Daejeon','Daegu','Ulsan','Busan','Gwangju']

lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] # 위도
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85] # 경도

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995] # 2017년 통계청 자료 - 인구밀도

size = np.array(pop_den)
colors = ['r','g','b','c','m','k','y']

plt.scatter(lon,lat, s=size, c=colors, alpha=0.5)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('2017')

for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
```

## 막대 그래프
막대의 높이로 값을 나타내어 여러 항목의 수량이 많고 적음을 한눈에 파악해 데이터를 서로 비교할 때 주로 이용

`plt.bar(x, height [,width=width_f, color=colors, tick_label=tick_labels, align='center', label=labels])`

- 가로 막대 그래프 그리는 방법은 bar를 barh로 작성하면 됨, 다른 점은 barh()는 width 옵션을 이용할 수 없다는 것


- height : 시각화하고자 하는 막대 그래프의 데이터


- x : height와 길이가 일치하는 데이터로 x축에 표시될 위치를 지정, 순서만 지정하므로 0부터 시작해서 height의 길이만큼 1씩 증가하는 값을 가짐


- width : 0과 1사이의 실수를 지정해 막대의 폭을 조절(기본값 0.8)


- tick_label : 문자열 또는 문자열 리스트를 입력해 막대 그래프 각각의 이름 지정(기본값 숫자)


- align : 막대 그래프의 위치
    - center : 가운데(기본값)
    - edge : 한쪽으로 치우지게
    

- label : 범례에 사용될 문자열

```python
# 헬스클럽 예제

member_ids = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]

n_data = len(member_ids)
index = np.arange(n_data)
colors = ['r', 'g', 'b', 'm']

plt.bar(index, before_ex, tick_label=member_ids, width=0.6, color=colors)

plt.show()
```

```python
# 헬스클럽 예제(가로 막대 그래프)

member_ids = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]

n_data = len(member_ids)
index = np.arange(n_data)
colors = ['r', 'g', 'b', 'm']

plt.barh(index, before_ex, tick_label=member_ids, color=colors)

plt.show()
```

```python
# 헬스클럽 예제

member_ids = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]

n_data = len(member_ids)
index = np.arange(n_data)
labels=['before','after']
barWidth = 0.4


plt.bar(index, before_ex,
        color='r', width=0.4, align='edge',
       label='before')
plt.bar(index+barWidth, after_ex, width=0.4,
        color='c', align='edge', label='after')

plt.xticks(index+barWidth, member_ids)
plt.xlabel('Member ID')
plt.ylabel('Excercise')
plt.title('Excercise Before & After')
plt.legend()


plt.show()
```

## 파이그래프
- 원 안에 데이터의 각 항목이 차지하는 비율만큼 부채꼴의 크기를 갖는 영역으로 이뤄짐
- 부채꼴 부분의 크기는 각 항목 크기에 비례함
- 전체 데이터에서 각 항목이 차지한 비율을 비교할 때 주로 이용


`plt.pie(x [, labels=label_seq, autopct='비율표시형식', shadow=False(기본값)/True, explode = explode_seq, counterclock=True(기본값)/False, startangle=0(기본값)`


- x : 배열, 시퀀스 형태로 x의 각 요소가 전체에서 차지하는 비율을 계산하고 그 비율에 맞게 부채꼴 부분의 크기를 결정해서 그래프를 그림

- labels : x의 요소 수와 같은 문자열 시퀀스를 지정해 각 데이터에 문자열을 표시(이름붙이기)

- autopct : 각 부채꼴 부분에 항목의 비율이 표시되는 숫자의 형식을 지정. %0.0f를 입력할 시 정수만 표시, %0.1f 입력시 소수점 첫째 자리까지 표시

- shadow : 그림자 효과 지정

- explode : 특정 부채꼴 영역을 강조하고 싶을 때 사용. x 데이터의 항목 수와 같은 시퀀스로 지정.

- counterclock : x 데이터에서 부채꼴을 그리는 순서의 방향. True는 반시계방향(기본값), False는 시계방향.

- startangle : 제일 처음 부채꼴이 그려지는 각도로 x축 중심으로 반시계방향으로 증가.


`plt.figure(figsize=(w,h))`

- w, h 값을 생략할 시 기본값은 (6,4)가 됨


```python
fruit = ['apple', 'melon', 'strawberry', 'pear', 'watermelon']
result = [7, 6, 3, 2, 2]

plt.pie(result, labels=fruit, autopct='%0.1f%%')
plt.show()
```

```python
# apple 데이터 강조하기
explode_data = (0.5, 0, 0, 0, 0)

plt.pie(result, labels=fruit, autopct='%0.1f%%', startangle=90, explode=explode_data)
plt.show()
```
