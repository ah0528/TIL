* **키가 되는 기준열 이름이 두 데이터 프레임에서 다르게 나타나는 경우**

left_on, right_on 인수를 사용해서 기준열을 명시해야 함

```python
df1=pd.DataFrame({
    '이름' :['영희','철수','철수'],
    '성적' :[90,80,80]
})
df2 = pd.DataFrame({
    '성명' :['영희','영희','철수'],
    '성적2':[100,80,90]
})
df1
df2
```

```python
# concat 연결
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'E': ['C4', 'C5', 'C6', 'C7'],
                    'F': ['D4', 'D5', 'D6', 'D7']},
                   index=[0, 1, 2, 3])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[1,2,3,4])
df1
df2
df3
```

연습문제

* 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과 후반기(7월-12월)실적을 나타내는 데이터 프레임을 작성한 후 합친다(단순병합)

* 실적 정보는 "매출","비용","이익"으로 이루어진다.

* 이익=매출-비용 : 가공필드로 생성

* 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

```python
month = [str(i)+'월' for i in range(1,13)]
month
```

```python
# 매출, 비용 data random  생성
np.random.seed(3)

# 1~6월의 매출, 비용 데이터 생성(딕셔너리)
data={}
for i in month[0:6]:
    data[i] = np.random.randint(1,100,2)
    
data
```

```python
# 데이터프레임 생성

result = pd.DataFrame(data, index=['매출','비용'])
result
```

```python
# '이익' 행 생성 : 매출 - 비용
result.loc['이익'] = result.loc['매출']-result.loc['비용']
result
```

```python
# 하반기(7~12월) 실적 데이터프레임 생성
# 매출, 비용 data random  생성
np.random.seed(2)
data={}
# 데이터프레임 생성
for i in month[6:13]:
    data[i] = np.random.randint(1,100,2)

result2 = pd.DataFrame(data, index=['매출', '비용'])
result2
# 가공 행 이익 생성
result2.loc['이익'] = result2.loc['매출'] - result2.loc['비용']
result2
```

```python
result3 = pd.concat([result,result2], axis=1)
result3
```

```python
result3['총실적'] = result3.sum(axis=1).astype(int)
result3
```
